Criao do modelo

```python
model = keras.models.Sequential()

model.add(keras.layers.InputLayer(input_shape=(28,28))) #Os dados vo chegar dessa forma
model.add(keras.layers.Rescaling(scale=1./255)) #Vo sofrer essa transformao
model.add(keras.layers.Flatten()) #E vai achatar
model.add(keras.layers.Dense(128, activation="relu")) #Criando a camada central
model.add(keras.layers.Dense(10,activation="softmax")) #Camada de sada

model.summary()
```

Compilar o modelo Keras

```python
opt = keras.optimizers.Adam()
loss = keras.losses.SparseCategoricalCrossentropy()
metric = keras.metrics.SparseCategoricalAccuracy()

model.compile(optimizer= opt, loss=loss, metrics=[metric])
```

Fit do modelo

```python
early_stop = keras.callbacks.EarlyStopping(
    patience=5,
    verbose=1,
    min_delta=1E-4,
)

history = model.fit(
    x_train,
    y_train,
    epochs=100,
    batch_size=128,
    validation_split=0.2,
    callbacks= [early_stop],
)
```

Evaluate do modelo

```python
model.evaluate(x_test, y_test)
```

Predict do modelo

```python
plt.plot(history.history["loss"], label="loss")
plt.plot(history.history["val_loss"], label="val_loss")
plt.legend()
plt.show()
```

```python
plt.plot(history.history["sparse_categorical_accuracy"], label="sparse_categorical_accuracy")
plt.plot(history.history["val_sparse_categorical_accuracy"], label="val_sparse_categorical_accuracy")
plt.legend()
plt.show()
```

```python
fig, axs = plt.subplots(4,5, figsize=(17,6)) #4  o nmero de linhas e 5 o nmero de colunas

for i, ax in enumerate(axs.flat):
    ax.imshow(x_test[i], cmap='gray')
    pred = model.predict(np.expand_dims(x_test[i], axis=0), verbose=0)[0]
    ax.set_title(f"verdade: {y_test[i]}", pad=15) #pad  para espaar
    ax.axis("off")
    ax.text(
        0,
        -2,
        f"predio: {pred.argmax()}-{pred.max():.1%}",
        color="green" if pred.argmax() == y_test[i] else "red"
    )
fig.subplots_adjust(hspace=0.60) 
plt.show()
```

```python
from pathlib import Path

test_images_dir = Path(r"C:\Users\User\Desktop\Cincia de Dados\Curso Hashtag\Deep learning\test_images")

fig, axs = plt.subplots(2, 5, figsize=(10,5))

all_pred = {}

for i, (img, ax) in enumerate(zip(sorted(test_images_dir.glob("*.png")), axs.flat)):
    name = img.name.split(".")[0] #O [0]  por conta do batch que veio a partir do expand.dims
    #
    img = keras.preprocessing.image.load_img(
        img, target_size=(28,28), color_mode="grayscale"
    )
    #
    img_array = keras.preprocessing.image.img_to_array(img)
    img_array = 255 - img_array
    img_array = tf.expand_dims(img_array, 0)
    
    pred = model.predict(img_array, verbose=0)
    all_pred[name] = pred
    
    ax.imshow(img_array[0], cmap='gray')
    ax.set_title(f"verdade: {name}", pad=15) #pad  para espaar
    ax.axis("off")
    ax.text(
        0,
        -2,
        f"predio: {pred.argmax()}-{pred.max():.1%}",
        color="green" if str(pred.argmax()) == name else "red"
    )
fig.subplots_adjust(hspace=0.60) 
plt.show()
```
